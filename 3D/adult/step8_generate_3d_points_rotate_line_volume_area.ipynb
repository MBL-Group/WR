```python
import numpy as np
import cv2
import os


images_folder_with_fix_tail = "./N2-adults-50um-SRx5-fix-tail-crop-images-correct-angle-folders"
images_folder = './N2-adults-50um-SRx5-crop-images-correct-angle-folders'

seg_folder_with_fix_tail = "./N2-adults-50um-SRx5-fix-tail-crop-seg-correct-angle-folders"
seg_folder = './N2-adults-50um-SRx5-crop-seg-correct-angle-folders'


sub_folders = os.listdir(images_folder)

print(sub_folders)

pointcloud_folder = "N2-adults-50um-SRx5-3D-pointcloud"
```

```python
import math
import open3d
from copy import deepcopy

fx_scale = 5.0 

for sub_folder in sub_folders:

    images_list_dir = images_folder+'/'+sub_folder
    seg_images_list_dir = seg_folder+'/'+sub_folder

    images_list = os.listdir(images_list_dir)
    seg_images_list = os.listdir(seg_images_list_dir)


    
    
    print(sub_folder)


    
    pcd_offset_file = open3d.io.read_point_cloud(pointcloud_folder+'/'+sub_folder+'/best_iou_after_offset.pcd')
    
    origin_pcd = deepcopy(pcd_offset_file)
    

    surface_area = 0
    volume = 0

    

    
    for j in range(2600):

        

        xyz_ppints = np.array(origin_pcd.points).astype(int)

        where_res=np.where(xyz_ppints[:,1] == j)

        xz_layer = xyz_ppints[where_res]

        xz_image = np.zeros((500,500))

        xz_data = []

        for point in xz_layer:
            x,y,z = point
            
            xz_data.append([x+239,z+239])

        if len(xz_data) >0:

            xz_data = np.array(xz_data).reshape(-1,1,2)

            approx1 = cv2.approxPolyDP(xz_data,3,False)

            res_hull=cv2.convexHull(approx1)

            xz_image  =cv2.polylines(xz_image,[res_hull],True,(255,255,255),1)

            
            area = cv2.contourArea(res_hull)/(fx_scale*fx_scale) 
            volume += (area*(1/fx_scale))
            

            
            perimeter = cv2.arcLength(res_hull, True)/fx_scale
            surface_area += (perimeter * (1/fx_scale))
            


            
        

        


    print('surface area = ', surface_area)
    print('volume = ', volume)
            
```
